```python
 
#We are going to use our air passenger dataset, so upload this to COLAB,
# then load into pandas
# The file is available at the following link
ruta_completa = 'output2_2023_eit171.csv'

data = pd.read_csv(ruta_completa)

# Quickly check the data looks like we expect
print(data.head())
```

```python
import pandas as pd

# Supongamos que tienes un DataFrame llamado "data" y la columna "Date" contiene las fechas en el formato "01/01/2023 01:00"

# Convierte la columna "Date" a tipo de dato datetime especificando el formato
data['DateFormatted'] = pd.to_datetime(data['date'], format='%d/%m/%Y %H:%M')

# Establece la columna "DateFormatted" como el índice de tiempo
data.set_index('DateFormatted', inplace=True)

# Ahora "DateFormatted" se convierte en el índice de tiempo con fechas válidas
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 12))  


plt.subplot(4, 4, 1)
plt.plot(data['entropy'])
plt.title('Entropy')

plt.subplot(4, 4, 2)
plt.plot(data['fractal_dimension'])
plt.title('Fractal Dimension')

plt.subplot(4, 4, 3)
plt.plot(data['mean_intensity'])
plt.title('Mean Intensity')

plt.subplot(4, 4, 4)
plt.plot(data['standard_deviation'])
plt.title('Standard Deviation')

plt.subplot(4, 4, 5)
plt.plot(data['uniformity'])
plt.title('Uniformity')

plt.subplot(4, 4, 6)
plt.plot(data['skewness'])
plt.title('Skewness')

plt.subplot(4, 4, 7)
plt.plot(data['kurtosis'])
plt.title('Kurtosis')

plt.subplot(4, 4, 8)
plt.plot(data['relative_smoothness'])
plt.title('Relative Smoothness')

plt.subplot(4, 4, 9)
plt.plot(data['taruma_contrast'])
plt.title('Taruma Contrast')

plt.subplot(4, 4, 10)
plt.plot(data['taruma_directionality'])
plt.title('Taruma Directionality')

plt.subplot(4, 4, 11)
plt.plot(data['taruma_coarseness'])
plt.title('Taruma Coarseness')

plt.subplot(4, 4, 12)
plt.plot(data['taruma_linelikeness'])
plt.title('Taruma Linelikeness')

plt.subplot(4, 4, 13)
plt.plot(data['taruma_roughness'])
plt.title('Taruma Roughness')

plt.subplot(4, 4, 14)
plt.plot(data['taruma_regularity'])
plt.title('Taruma Regularity')

plt.tight_layout()


plt.show()
```
